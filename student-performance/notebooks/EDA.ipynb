# EDA (Exploratory Data Analysis)

This notebook is dedicated to loading, cleaning, and exploring the dataset. It focuses on understanding the structure, quality, and basic patterns in the data. For advanced or presentation-ready visualizations, see the `Visualisation.ipynb` notebook.

Importing Libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading the data:

```python
df = pd.read_csv('../data/StudentsPerformance.csv')
```

Getting a basic understanding of our data:

```python
# Looking at the first few rows of the dataframe
print(df.head())

# Looking at the data types and non-null counts
print(df.info())

# Summary statistics for numerical columns
print(df.describe())

# Summary statistics for categorical columns
print(df.describe(include=['object']))

# Checking the shape of the dataframe
print(f"Dataframe shape: {df.shape}")
```

Checking for missing values:

```python
# Checking for missing values in each column
print("Missing values in each column:")
df.isnull().sum()

# Visualizing missing values using a heatmap - this won't work well with no missing values!
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

Analysing Individual Columns:

```python
# Distribution of numerical columns
num_cols = ['math score', 'reading score', 'writing score']

# Individual histograms for numerical columns
for col in num_cols:
    plt.figure(figsize=(6, 4))
    plt.hist(df[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(f"{col} (Score)")
    plt.ylabel("Number of Students")
    plt.tight_layout()
    plt.show()

# Distribution of categorical columns
cat_cols = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

# Individual count plots for categorical columns
for col in cat_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col, palette='pastel')
    plt.title(f"Distribution of {col}")
    plt.xlabel(f"{col} (Categories)")
    plt.ylabel("Number of Students")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

Relationships between features:

```python
# Numerical vs Numerical
# Scatter plots for numerical columns
sns.pairplot(df[num_cols], diag_kind='kde', plot_kws={'alpha':0.6})
plt.suptitle('Pairwise Scatter Plots of Numerical Columns', y=1.02)
plt.show()

# Heatmap for correlation matrix
corr = df[num_cols].corr()
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()

# Categorical vs Categorical
# Crosstabs and heatmaps for categorical columns
ct = pd.crosstab(df['gender'], df['lunch'])
sns.heatmap(ct, annot=True, cmap='Blues', fmt='d')
plt.title('Gender vs Lunch Crosstab')
plt.show()

# Numerical vs Categorical
# Box plots for numerical vs categorical columns
plt.figure(figsize=(12, 4))
for i, col in enumerate(num_cols):
    plt.subplot(1, 3, i+1)
    sns.boxplot(data=df, x='gender', y=col, palette='Set2')
    plt.title(f'{col} by Gender')
plt.tight_layout()
plt.show()

# Violin plots for numerical vs categorical columns
plt.figure(figsize=(12, 4))
for i, col in enumerate(num_cols):
    plt.subplot(1, 3, i+1)
    sns.violinplot(data=df, x='test preparation course', y=col, palette='Set3')
    plt.title(f'{col} by Test Preparation Course')
plt.tight_layout()
plt.show()
```

Identifying Outliers:

```python
# Finding the IQR for numerical columns
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1
print('IQR values:')
print(IQR)

# Using the IQR to identify outliers
outliers = ((df[num_cols] < (Q1 - 1.5 * IQR)) | (df[num_cols] > (Q3 + 1.5 * IQR))).sum()
print('Number of outliers in each numerical column:')
print(outliers)

# Visualizing outliers using box plots
plt.figure(figsize=(12, 4))
for i, col in enumerate(num_cols):
    plt.subplot(1, 3, i+1)
    sns.boxplot(data=df, y=col, color='lightcoral')
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

Detecting Correlations:

```python
# Find the correlation matrix
corr_matrix = df[num_cols].corr()
print('Correlation matrix:')
print(corr_matrix)

# Heatmap of the correlation matrix
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

Feature Distributions across categories:

```python
# Pair plot for a target column
sns.pairplot(df, hue='gender', vars=num_cols, palette='husl')
plt.suptitle('Pairplot of Scores by Gender', y=1.02)
plt.show()
```
